### Importando Modelo - Phi-3-mini

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

torch.random.manual_seed(0)

model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3.5-mini-instruct",
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=True,
)
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3.5-mini-instruct")
```

### Relatorio de uma empresa ficticia

```python
problema_empresa = """
Relatório de Desempenho Anual - TechNova Eletrônicos
Período: Janeiro a Dezembro de 2023

Visão Geral de Vendas

Receita total: R$ 150 milhões (aumento de 15% em relação ao ano anterior)
Unidades vendidas: 500.000 (aumento de 12% em relação ao ano anterior)


Desempenho por Categoria de Produto
a) Smartphones: 45% da receita
b) Tablets: 25% da receita
c) Laptops: 20% da receita
d) Acessórios: 10% da receita
Satisfação do Cliente

Índice geral de satisfação: 8.2/10 (aumento de 0.3 pontos em relação ao ano anterior)
NPS (Net Promoter Score): 65 (aumento de 5 pontos)


Desempenho de Vendas por Região

Sudeste: 40%
Nordeste: 25%
Sul: 20%
Centro-Oeste: 10%
Norte: 5%


Análise de Produtos
a) Smartphones

Vendas: 225.000 unidades (aumento de 20% em relação ao ano anterior)
Satisfação do cliente: 8.5/10
b) Tablets
Vendas: 125.000 unidades (aumento de 15% em relação ao ano anterior)
Satisfação do cliente: 8.3/10
c) Laptops
Vendas: 100.000 unidades (aumento de 5% em relação ao ano anterior)
Satisfação do cliente: 7.6/10
d) Acessórios
Vendas: 50.000 unidades (diminuição de 2% em relação ao ano anterior)
Satisfação do cliente: 8.4/10


Reclamações e Devoluções

Taxa geral de devolução: 3% (sem alteração em relação ao ano anterior)
Principais motivos de reclamação:

Problemas de software (35% das reclamações)
Defeitos de hardware (30% das reclamações)
Expectativas não atendidas (20% das reclamações)
Problemas de entrega (15% das reclamações)




Análise de Mídias Sociais

Menções positivas: 70%
Menções neutras: 20%
Menções negativas: 10%


Tendências de Mercado

Aumento na demanda por dispositivos com maior duração de bateria
Crescente interesse em produtos eco-friendly
Preferência por designs minimalistas


Iniciativas de Marketing

Campanhas de mídia social: ROI de 250%
Programa de fidelidade: 100.000 novos membros


Metas para o Próximo Ano

Aumentar a receita total em 20%
Expandir a presença no mercado da região Norte
Lançar uma nova linha de produtos eco-friendly
"""

```

### Cria a prompt e gera o resumo do documento

```python
messages = [
    {"role": "user", "content": "Resuma o seguinte documento e diminua seu tamanho total, mantenha a coesão, os dados e as estatisticas: " + problema_empresa}
]

pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
)

generation_args = {
    "max_new_tokens": 1024,
    "return_full_text": False,
    "temperature": 0.0,
    "do_sample": False,
}

output = pipe(messages, **generation_args)
print(output[0]['generated_text'])

```

```python
print(len(output[0]['generated_text']), len(problema_empresa))
```
