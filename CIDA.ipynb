### Importando Modelo - Phi-3-mini

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

torch.random.manual_seed(0)

model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-128k-instruct",
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=True,
)
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-128k-instruct")
```

### Relatorio de uma empresa ficticia

```python
problema_empresa = """

Relatório Semestral - Empresa X
Período: 01/01/2024 - 30/06/2024

Introdução

O presente relatório tem como objetivo apresentar os resultados da Empresa X durante o primeiro semestre de 2024, abrangendo os aspectos financeiros, operacionais, comerciais e de mercado.

Desempenho Financeiro

No primeiro semestre de 2024, a Empresa X registrou um lucro líquido de R$ 12,5 milhões, representando um crescimento de 15% em relação ao mesmo período do ano anterior. Esse resultado positivo foi impulsionado por diversos fatores, como:

Aumento da receita líquida em 10%, alcançando R$ 85 milhões, devido principalmente ao crescimento das vendas em 12% para os produtos X, Y e Z.
Redução dos custos operacionais em 8%, totalizando R$ 55 milhões, em decorrência de medidas de otimização da cadeia de suprimentos e implementação de programas de eficiência energética.
Melhoria da margem líquida para 14,7%, demonstrando a capacidade da empresa de gerar lucro a partir da receita líquida.
Desempenho Operacional

A Empresa X manteve um bom desempenho operacional durante o semestre, com os seguintes destaques:

Produção física em linha com as metas estabelecidas, alcançando 10.000 unidades.
Nível de inventário adequado, equivalente a 30 dias de estoque.
Alta taxa de aproveitamento da capacidade instalada, de 85%.
Baixo índice de retrabalho, de apenas 2%.
Desempenho Comercial

A Empresa X obteve um bom desempenho comercial no semestre, com os seguintes resultados:

Aumento da carteira de clientes em 5%, alcançando 2.500 clientes.
Expansão da presença em novos mercados, com a abertura de 2 novas filiais.
Lançamento de 3 novos produtos, bem recebidos pelo mercado.
Forte atuação no marketing digital, com um aumento de 20% no número de seguidores nas redes sociais.
Desafios e Perspectivas

Apesar do bom desempenho geral, a Empresa X enfrenta alguns desafios, como:

Aumento da concorrência no mercado, com a entrada de novos players e o lançamento de produtos similares por empresas já estabelecidas.
Pressão nos preços de venda, devido à retração da economia e à queda do poder de compra dos consumidores.
Instabilidade do cenário político e econômico, que pode afetar a demanda por produtos da empresa.
Diante desses desafios, a Empresa X tem as seguintes perspectivas para o segundo semestre de 2024:

Manter o crescimento da receita líquida, através da expansão da carteira de clientes e do lançamento de novos produtos.
Reduzir custos operacionais, buscando otimizar ainda mais a cadeia de suprimentos e implementar novas medidas de eficiência.
Investir em inovação, para desenvolver produtos e serviços diferenciados que atendam às necessidades dos clientes.
Fortalecer a presença no mercado digital, expandindo a atuação nas redes sociais e no e-commerce.
Conclusão

A Empresa X concluiu o primeiro semestre de 2024 com resultados positivos, demonstrando solidez financeira, eficiência operacional e bom desempenho comercial. No entanto, a empresa está ciente dos desafios que enfrenta e está tomando medidas para superá-los, buscando o crescimento sustentável e a longo prazo.
"""

```

### Cria a prompt e gera o resumo do documento

```python
messages = [
    {"role": "user", "content": "Resuma o seguinte documento, porém mantenha os dados e as estatisticas: " + problema_empresa}
]

pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
)

generation_args = {
    "max_new_tokens": 500,
    "return_full_text": False,
    "temperature": 0.0,
    "do_sample": False,
}

output = pipe(messages, **generation_args)
print(output[0]['generated_text'])

```
